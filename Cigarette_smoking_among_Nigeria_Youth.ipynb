```python
# Import Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Loading Data
df = pd.read_csv('smoking_data.csv')
```

```python
# Showing the first rows of the dataframe

df.head(6)
```

```python
# Handling missing values and replacing the NaN for categorical column

df['Type of Tobacco Product'] = df['Type of Tobacco Product'].replace(np.nan, 'None')
df['Reasons for Smoking'] = df['Reasons for Smoking'].replace(np.nan, 'Dont Smoke')
df= df.dropna()
```

```python
df['Previous Quit Attempts'] = df['Previous Quit Attempts'].astype(int)
df['Cost of Smoking per month'] = df['Cost of Smoking per month'].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Binanry Encoding for Gender, Awareness of Health Risks, Intent to Quit, Access to Cessation Support, Influence of Advertising


df['Gender_Encoded']= df['Gender'].map({'Male': 0, 'Female': 1})
df['Awareness_of_Health_Risks_Encoded']= df['Awareness of Health Risks'].map({'Yes': 1, 'No': 0})
df['Intent_to_Quit_Encoded']= df['Intent to Quit'].map({'Yes': 1, 'No': 0})
df['Access_to_Cessation_Support_Encoded']= df['Access to Cessation Support'].map({'Yes': 1, 'No': 0})
df['Influence_of_Advertising_Encoded']= df['Influence of Advertising'].map({'High': 2, 'Medium': 1, 'Low': 0})
df['Employment_Status_Encoded']= df['Employment Status'].map({'Unemployed': 2, 'Employed': 1, 'Student': 0 })

#Label encoding for Education Level, Perceived Health Status, Economic Status

label_encoder = LabelEncoder()

df['Education_Level_Encoded'] = label_encoder.fit_transform(df['Education Level'])
df['Perceived_Health_Status_Encoded'] = label_encoder.fit_transform(df['Perceived Health Status'])
df['Economic_Status_Encoded'] = label_encoder.fit_transform(df['Economic Status'])

# One-hot encoding for Location, Employment Status, Smoking Status, Type of Tobacco product, Reasons for Smoking

df_one_hot_encoding = pd.get_dummies(df,columns = ['Location', 'Employment Status', 'Smoking Status',
                                           'Type of Tobacco Product', 'Reasons for Smoking'])


df_encoded= pd.concat([df, df_one_hot_encoding], axis=1)


df_encoded.head()

```

```python
# Distribution of Age

plt.figure(figsize= (5,6))
sns.histplot(df['Age'], bins= 10,kde= True)
plt.title('Age Distribution of Smokers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

print(''' Insights
The histogram and kernel density estimate of the age distribution of smokers reveal a bimodal pattern, indicating two primary age groups approximately
18 - 19 and 23 - 24 years old where smoking is most prevalent.''')
```

```python
# Bar chart of smoking status

plt.figure(figsize=(5,5))
sns.countplot(x= 'Smoking Status', data= df)
plt.title('Distribution of Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()

print('''The bar chart indicates that the majority of individuals in the dataset are current smokers
followed by former smokers, and the smallest group being those who have never smoked.''')
```

```python
# Bar chart of smoking status

plt.figure(figsize=(5,5))
sns.countplot(x= 'Smoking Status_Current', data= df_encoded)
plt.title('Bar chart of Smoking Status')
plt.xlabel(' Current Smoking Status')
plt.ylabel('Count')
plt.show()

print('''The bar chart indicates that in the sampled population, there are more individuals who do not (Currently Smoke) compared to those
 who Smokes''')
```

```python
# Histogram of Frequency of Smoking

plt.figure(figsize=(7,4))
sns.histplot(df['Frequency of Smoking per day'], bins=10, kde= True)
plt.title('Distribution of Frequency of Smoking')
plt.xlabel('Frequency of Smoking (per day)')
plt.ylabel('Frequency')
plt.show()

print('''The right-skewed data shows that there are substantially more people who either do not smoke or smoke up to once per day compared to those who 
smoke frequently''')
```

```python
# Distribution of Tobacco status

plt.figure(figsize=(5,5))
sns.countplot(x= 'Type of Tobacco Product', data= df)
plt.title('Tobacco Product')
plt.show()

print('''This analysis reveals that Cigarettes Smokers are most common among Nigerian Youth with the highest count followed by the Shisha and E-cigarettes
smokers''')
```

```python
# Reasons for Smoking

smoking_reasons = df['Reasons for Smoking'].value_counts()
plt.figure(figsize=(7,4))
smoking_reasons.plot(kind='pie')
plt.title('Reasons for Smoking')
plt.show()


print('''The pie chart highlights that addiction and stress relief are the primary reasons for Smoking among youth, with social influence and peer pressure
also contributing but to a lesser extent''')
```

```python
# Box plot of Age at Initiation by Smoking Status

plt.figure(figsize=(5,5))
sns.boxplot(x= 'Smoking Status', y= 'Age at Initiation', data= df)
plt.title('Age at Initiation by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Age at Initiation')
plt.show()

print('''The majority of smoking initiation occurs during the teenage years for both groups. For current smokers, the median age of smoking
initiation is indeed around 17 years. While for Fomer smokers the median initiation age is around 16 years.''')
```

```python
# Violin Plot Frequency of Smoking per_day by Smoking Status

plt.figure(figsize=(7,7))
sns.violinplot(x= 'Smoking Status', y='Frequency of Smoking per day', data=df)
plt.title('Distribution of Smoking per_day by Smoking Status')
plt.show()

print('''The smoking frequency among different statuses reveals distinct pattern. For current smokers, the median smoking frequency is
approximately 4 Cigarettes per day. while Former smokers have a median frequency of around 1 Cigarette per day.''')
               
```

```python
# Comparative Analysis of smoking rate by gender

plt.figure(figsize=(8,6))
sns.countplot(x='Gender',hue='Smoking Status', data=df)
plt.title('Smoking Status by Gender')
plt.legend(title='Smoking Status')
plt.show()

print('''The analysis reveals that both genders have a higher number of current smokers compared to former and never smokers''')
```


## Conclusion:

 - Reasons for smoking such as stress relief and addiction are prevalent amont Nigerian Youths, highlighting the importance of addressing social and pyschological factors in smoking cessation efforts.

 - Addiction Levels vary among different smoking statuses, with current smokers showing higher frequencies of smoking compared to former and never smokers.

 - The analysis reveals that Age and Gender have significant relationship, showing that Early initiation of smoking especially among females, is correlated with higher smoking frequencies, and older males (ages 22-24) show more consistent smoking habits, with a notable proportion smoking around 4-5 times per day. Both education level and employment status plays roles in influencing smoking behaviour, with a high rate of cigratte smokers compare to other Tobacco products. This suggests targeted interventions might be necessary for younger age groups to prevent early initiation.

 - Different locations exhibit different primary reasons for smoking. Kano, with its high addiction rates, may need different strategies compared to Kaduna, where stress relief is a significant factor.

 - Students exhibit high counts across various reasons, indicating that educational settings might be critical for smoking smoking prevention programs. Unemployment seems to correlate strongly with addiction, suggesting economic factors play a role in smoking behaviours.

 - There's a clear correlation between perceived poorer health and higher smoking frequencies. Public health campaigns could emphasize the health risks associated with frequent smoking to improve health perceptions and reduce smoking rates.

 - Since females tend to start smoking earlier and at higher frequencies, gender-specific interventions could be more effective.
